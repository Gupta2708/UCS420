Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data, clean and process it, perform statistical analysis, and generate visual insights to help the business make informed decisions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

roll_number = 102316130
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)

print("First 5 rows of the DataFrame:")
print(df.head())

print("\nSummary Statistics:")
print(df.describe())

total_sales_category = df[categories].sum()
total_sales_month = df[categories].sum(axis=1)

print("\nTotal Sales per Category:")
print(total_sales_category)

print("\nTotal Sales per Month:")
print(total_sales_month)

sales_growth = df[categories].pct_change().mean() * 100

print("\nAverage Sales Growth (%) between Consecutive Months for Each Category:")
print(sales_growth)

df['Total Sales'] = df[categories].sum(axis=1)
df['Growth Rate (%)'] = df['Total Sales'].pct_change() * 100

df['Electronics (After Discount)'] = df['Electronics'] * 0.9

plt.figure(figsize=(12, 7))
for category in categories:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.title('Monthly Sales Trends per Category', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales Units', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category', fontsize=16)
plt.ylabel('Sales Units', fontsize=14)
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

print("Absolute Values:")
print(np.abs(array))

print("\nPercentiles (Flattened Array):")
for perc in [25, 50, 75]:
    print(f"{perc}th Percentile:", np.percentile(array, perc))

print("\nPercentiles (Column-wise):")
for perc in [25, 50, 75]:
    print(f"{perc}th Percentile:", np.percentile(array, perc, axis=0))

print("\nPercentiles (Row-wise):")
for perc in [25, 50, 75]:
    print(f"{perc}th Percentile:", np.percentile(array, perc, axis=1))

print("\nMean:")
print("Flattened Array:", np.mean(array))
print("Column-wise:", np.mean(array, axis=0))
print("Row-wise:", np.mean(array, axis=1))

print("\nMedian:")
print("Flattened Array:", np.median(array))
print("Column-wise:", np.median(array, axis=0))
print("Row-wise:", np.median(array, axis=1))

print("\nStandard Deviation:")
print("Flattened Array:", np.std(array))
print("Column-wise:", np.std(array, axis=0))
print("Row-wise:", np.std(array, axis=1))
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

print("Floor Values:")
print(np.floor(a))

print("\nCeiling Values:")
print(np.ceil(a))

print("\nTruncated Values:")
print(np.trunc(a))

print("\nRounded Values:")
print(np.round(a))
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
lst = [1, 2, 3, 4, 5]

def swap_list(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

print("Swapped List:", swap_list(lst, 1, 3))
```

Q.5 Write a Python program to swap two elements in a set by converting it to a list first.

```python
s = {10, 20, 30, 40, 50}
def swap_set_elements(s, val1, val2):
    lst = list(s)
    idx1, idx2 = lst.index(val1), lst.index(val2)
    lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
    return set(lst)
print("Swapped Set:", swap_set_elements(s, 20, 40))
```
